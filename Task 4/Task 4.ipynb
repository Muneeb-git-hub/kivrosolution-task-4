```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# 1: load the  data set
df=pd.read_csv('student_scores_.csv')
```

```python
df
```

```python
# 2. Display dataset shape, column info, summary statistics


print("Dataset Shape:", df.shape)
print("\nColumn Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
```

```python

# 3. Handle missing values (check, then decide)


print("\nMissing Values:")
print(df.isnull().sum())
```

```python

# 4. Create Visualizations

#  Visualization 1: Line Plot (Hours vs Score) 
plt.figure(figsize=(7,4))
plt.plot(df["Hours_Studied"], df["Test_Score"], marker='o', color='green')
plt.title("Hours Studied vs Test Score")
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")
plt.grid(True)
plt.show()
# Observation: Test scores increase steadily as students study more hours.


#  Visualization 2: Scatter Plot (Attendance vs Score)
plt.figure(figsize=(7,4))
plt.scatter(df["Attendance_Percentage"], df["Test_Score"], color='blue')
plt.title("Attendance vs Test Score")
plt.xlabel("Attendance Percentage")
plt.ylabel("Test Score")
plt.grid(True)
plt.show()
# Observation: Higher attendance is associated with higher test scores.


#  Visualization 3: Histogram of Test Scores 
plt.figure(figsize=(7,4))
plt.hist(df["Test_Score"], bins=5, color='orange', edgecolor='black')
plt.title("Distribution of Test Scores")
plt.xlabel("Test Score")
plt.ylabel("Frequency")
plt.show()
# Observation: Most test scores fall between 60 and 80, showing a clustered distribution.


# Visualization 4: Bar Chart (Student ID vs Score) 
plt.figure(figsize=(8,4))
plt.bar(df["Student_ID"], df["Test_Score"], color='purple')
plt.title("Test Score per Student")
plt.xlabel("Student ID")
plt.ylabel("Test Score")
plt.show()
# Observation: Scores gradually rise from Student 1 to Student 10, reflecting consistent improvement.
```
